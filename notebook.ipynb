```python
# # from mysql import connector
# import sys
# !{sys.executable} -m pip install pandas




import mysql.connector
from dotenv import load_dotenv
import os 
import pandas as pd
```

```python
load_dotenv()

user = os.getenv('USER')
password = os.getenv('PASSWORD')
host = os.getenv('HOST')

database = 'swiftmarket'
```

```python
connection = mysql.connector.connect(
                 user = user,
                 password = password,
                 host = host,
                 database = database
)

cursor = connection.cursor()
```

```python
connection.is_connected()
```

```python
query = '''show tables'''
cursor.execute(query)
rows = cursor.fetchall()
for row in rows :
    print(row)
```

```python
query = '''select * from salesdata limit 5'''
cursor.execute(query)
rows = cursor.fetchall()
for row in rows :
    print(row)
```

```python
cursor.column_names
```

```python
pd.DataFrame(data=rows , columns=cursor.column_names)
```

```python
# Func. 
def read_query(query):
    cursor.execute(query)
    rows = cursor.fetchall()
    return pd.DataFrame(data = rows, columns=cursor.column_names)
```

```python
query = '''select * from customers limit 5'''
read_query(query)
```

```python
query = '''select * from salesdata limit 5'''
read_query(query)
```

```python
query = '''
        select c.firstname, c.lastname, sum(s.totalprice) AS TOTALPRICE
        from customers as c 
        inner join salesdata as s on 
        c.customerid = s.customerid
        group by c.firstname, c.lastname
        ORDER BY TOTALPRICE DESC
        LIMIT 10

'''
```

```python
read_query(query)
```
